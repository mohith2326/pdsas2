```python
#removing teh missing values and relacing them with mean,median or mode.question1)
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/train.csv')
missing_values = data.isnull().sum()
num_cols = ['Year','Kilometers_Driven','Seats','Price',]
for col in num_cols:
    if data[col].dtype in ['float64', 'int64']:
        data[col].fillna(data[col].mean(), inplace=True)
cat_cols = ['Name','Location','Fuel_Type', 'Transmission','Owner_Type','Mileage', 'Engine','Power','New_Price']
for col in cat_cols:
    if data[col].dtype == 'object':
        data[col].fillna(data[col].mode()[0], inplace=True)
data.dropna(subset=['Price'], inplace=True)
missing_values_after = data.isnull().sum()
data.to_csv('cleaned_data.csv', index=False)
cleaned_data = pd.read_csv('cleaned_data.csv')
cleaned_data





```

```python
#removing the str and keeping only int.remove values like cc,bhp,lakh and keep the numeric values as it is question2)
import pandas as pd
import re
def extract_numeric(text):
    matches = re.findall(r"[-+]?\d*\.\d+|\d+", str(text))
    if matches:
        return float(matches[0])
    return None
def clean_and_convert_columns(data, columns_to_clean):
    for col in columns_to_clean:
        data[col] = data[col].apply(extract_numeric)
    data['Engine'] = data['Engine'].astype(int)

columns_to_clean = ['Mileage', 'Engine', 'Power', 'New_Price']

clean_and_convert_columns(cleaned_data, columns_to_clean)

cleaned_data.head()
```
